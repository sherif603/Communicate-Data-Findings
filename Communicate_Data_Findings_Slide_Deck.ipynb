**Dataset**

```python
# The distribution of the trip duration over day hours
# Use seaborn.color_palette() function to return a list of color or continuous colormap defining a palette.
base_color = sb.color_palette('colorblind')[0]
# Use seaborn.set_style() function to set aesthetic style of the plots
sb.set_style('darkgrid')
# Use matplotlib.figure() function to control the default spacing of the subplots and the top level container of the subplots,
# and the top level container for all plot elements.
plt.figure(figsize=[ 11.69, 8.27])
# Use matplotlib.pyplot.title() to set a title for the axes.
plt.title('Distribution of Trip Duration over day hours')
# Use seaborn.countplot() function to show the counts of observations in each categorical bin using bars.
sb.countplot(data=df, x='start_time_hour', color=base_color);
# Use matplotlib.pyplot.xlabel to set the label for the x-axis.
plt.xlabel('Trip Start Hour of Day');
# Use matplotlib.pyplot.ylabel to set the label for the y-axis.
plt.ylabel('Count');
```

**Observation**

```python
# Duratuion Trip over the day of the week
# Use seaborn.color_palette() function to return a list of color or continuous colormap defining a palette.
default_color = sb.color_palette()[0]
dayofweek_order = df['start_time_dayofweek'].value_counts().index
# Use matplotlib.figure() function to control the default spacing of the subplots and the top level container of the subplots,
# and the top level container for all plot elements.
plt.figure(figsize=[ 11.69, 8.27])
# Use seaborn.countplot()** function to show the counts of observations in each categorical bin using bars.
sb.countplot(data = df, x = 'start_time_dayofweek', color = default_color, order=dayofweek_order);
# Use matplotlib.pyplot.title to set a title for the axes.
plt.title('Frequency in Start Time, Day of Week');
# Use matplotlib.pyplot.xlabel to set the label for the x-axis.
plt.xlabel('Start Time, Day of Week');
```

**Observation**

- Thursday has the most bikers with counts of 33,712 of unique values compared to other days.
- However, Saturday has the lowest number of bikers with counts of 14,414 of unique values.

```python
# the Distribution of Start Time, Day of week
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
sorted_counts = df['start_time_dayofweek'].value_counts()
# Use matplotlib.figure() function to control the default spacing of the subplots and the top level container of the subplots,
# and the top level container for all plot elements.
plt.figure(figsize=[ 11.69, 8.27])
# Use matplotlib.pyplot.title() to set a title for the axes.
plt.title('Distribution of Start Time During the Days of the Week')
# Use matplotlib.pyplot.pie() function to plot a pie chart
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, autopct='%1.1f%%');
# Use matplotlib.pyplot.axis('square') to create a square plot with equal axes
plt.axis('square');
# Use matplotlib.pyplot.legend() function to place a legend on the axes 
plt.legend(title ="Day of Week",loc ="upper left",bbox_to_anchor =(1, 0, 0.5, 1));  
# Use matplotlib.pyplot.show() function to display all figures. 
plt.show();
```

**Observation**

- From the days of the week, Thursday has the most bikers, compared to other days, with a percentage of 19.3%.
- However, Saturday has the lowest number of bikers, with a percentage of 8.2%.

```python
# Ploting the distribution of bikers' gender

# Use matplotlib.pyplot.pie() function to plot a pie chart
sorted_counts = df['member_gender'].value_counts()
# Use matplotlib.figure() function to control the default spacing of the subplots and the top level container of the subplots,
# and the top level container for all plot elements.
plt.figure(figsize=[ 11.69, 8.27]);
# Use matplotlib.pyplot.title() to set a title for the axes.
plt.title('The Distribution of Gender');
# Use matplotlib.pyplot.pie() function to plot a pie chart
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, autopct='%1.1f%%');
# Use matplotlib.pyplot.axis('square') to create a square plot with equal axes
plt.axis('square');
# Use matplotlib.pyplot.legend() function to place a legend on the axes 
plt.legend(title ="Gender",loc ="upper left",bbox_to_anchor =(1, 0, 0.5, 1));  
# Use matplotlib.pyplot.show() function to display all figures. 
plt.show();
```

**Observation**

- The male bikers are about three times of female bikers.
- The value counts of "Male", "Female", and "Other" are 126311, 40098, and 3583 respectively.
- However, the percentage of "Male" is approximately 74%, "Female" is approximately 24%, and "Other" is approximately 2% off the "Gender" feature. 
- In other words,the number of times a data value occurs or the frequency of "Male" is greater than the frequency of "Female" and "Other" in the "Gender" feature.

```python
# Use matplotlib.figure() function to control the default spacing of the subplots and the top level container of the subplots,
# and the top level container for all plot elements.
plt.figure(figsize=[ 11.69, 8.27]);
plt.title('Plotting Heatmap of Correlation Matrix');
# Use seaborn.heatmap() function to plot rectangular data as a color-encoded matrix.
# Use Pandas dataframe.corr() is used to find the pairwise correlation of all columns in the dataframe.
# Any na values are automatically excluded. For any non-numeric data type columns in the dataframe it is ignored.
sb.heatmap(df.corr(), annot = True, fmt = '.3f', cmap = 'vlag_r', center = 0);
```

**Observation**

- The blue color with a positive number means positive correlation between the two variables, however, the red color with minus number means negative correlation between the two variables. The stronger the color, the larger the correlation magnitude.

- Start Station Latitute has the highest positive correlation with End Station Latitude.
- Start Station Longitude has the highest positive correlation with End Station Longitude.
- Duration Minutes has the highest positive correlation with Duration Seconds.

- Start Station Latitute has the highest negative correlation with End Station Longitude, followed by Start Station Longitude.
- Start Station Longitude has the highest negative correlation with End Station Latitude, followed by Start Station Latitude.
